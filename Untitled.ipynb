```python
import pybullet as p
import pybullet_data
```

```python
physicsClient = p.connect(p.GUI)#or p.DIRECT for non-graphical version
p.setAdditionalSearchPath(pybullet_data.getDataPath()) #used by loadURDF
p.setGravity(0,0,-10)
planeId = p.loadURDF("plane.urdf")
cubeStartPos = [0,0,1]
cubeStartOrientation = p.getQuaternionFromEuler([0,0,0])
```

```python
boxId = p.loadURDF("models/franka_gripper/franka/franka_gripper.urdf", cubeStartPos, cubeStartOrientation)
```

```python
p.getNumJoints(boxId)
p.getJointInfo(boxId,0)
p.getJointInfo(boxId,1)
p.getJointInfo(boxId,2)
p.getJointInfo(boxId,3)
p.getJointInfo(boxId,6)
p.getJointInfo(boxId,7)
```

```python
p.getLinkState(boxId,2)

```

```python
low = -10
high = 10
sliders = []
for i in range(4):
    slider = p.addUserDebugParameter('{}'.format(i), low, high, 0)
    sliders.append(slider)
```

```python
while True:
    positions = [p.readUserDebugParameter(slider) for slider in sliders]
    p.setJointMotorControlArray(boxId, 
                                [0,1,2,3], 
                                p.POSITION_CONTROL,
                                targetPositions=positions)
    pos, orn, _, _, _, _ = p.getLinkState(boxId, 4)
    print("pos ", pos)
    p.stepSimulation()
```

```python
p.disconnect()
```
